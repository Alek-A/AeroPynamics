# Potential Flows

```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Making a meshed grid

```python
N = 50                            # no. of points in each direction
x_start, x_end = -2.0, 2.0        # boundaries for x-dir.
y_start, y_end = -2.0, 2.0        # boundaries for y-dir.
x = np.linspace(x_start, x_end, N)    #
y = np.linspace(y_start, y_end, N)
X, Y = np.meshgrid(x,y)               # meshes grid
```

```python
def twoD_source(x0,y0, X,Y, m):
    R = np.sqrt((X-x0)**2 + (Y-y0)**2)
    u = m/2/np.pi*(X-x0)/((X-x0)**2 + (Y-y0)**2)
    v = m/2/np.pi*(Y-y0)/((X-x0)**2 + (Y-y0)**2)
    return u, v
```

```python
# Plotting streamlines
x_source = -0.8
y_source = 1.0
u, v = twoD_source(x_source,y_source, X,Y, 2)


size = 5
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
```
